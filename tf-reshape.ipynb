## **Initialization**

```
import tensorflow as tf
```

### **Main code**

### tf.reshape() changes the shape of a tensor without changing its data.

```
t1 = tf.constant([
    [1,2,3],
    [4,5,6]
])

t1.shape
```

```
tf.reshape(t1,[3,2])
```

### -1 automatically calculates the dimension needed to maintain total elements

```
tf.reshape(t1,[-1,1])
```
