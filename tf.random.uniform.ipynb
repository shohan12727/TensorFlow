### tf.random.uniform( )

```
import tensorflow as tf
```

```
t1 = tf.random.uniform(shape=(5,))
t1
```

## Can set min value and max value

```
t2 = tf.random.uniform(shape=(19,), minval=11, maxval=18)
print(t2)
```
